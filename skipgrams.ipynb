```python
import sklearn
```

```python
from sklearn.datasets import fetch_20newsgroups
from nltk.tokenize import sent_tokenize
```

```python
categories = ['rec.motorcycles', 'comp.graphics', 'sci.med']
corpus = fetch_20newsgroups(subset='train', categories=categories, shuffle=True, random_state=40,
                            remove=('headers', 'footers', 'quotes'))
```

```python
from nltk.util import skipgrams
len(corpus)
```

```python
len(corpus['data'])
```

```python
corpus['data']
```

```python
sentence_list = []
for post in corpus['data']:
    sent_tokenize_list = sent_tokenize(post)
    sentence_list.extend(sent_tokenize_list)
```

```python
from nltk.tokenize import word_tokenize
word_tokenize(sentence_list)
```

```python
word_list =[]
for sent in sentence_list:
    word_tokenize(sent)
    word_list.append(word_tokenize(sent))
```
